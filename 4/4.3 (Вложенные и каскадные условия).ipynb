Если пользователь захочет поделить на ноль, выведите текст «На ноль делить нельзя!» (без кавычек).

если смешать красный и синий, то получится фиолетовый;
если смешать красный и желтый, то получится оранжевый;

На числовой прямой даны два отрезка:[a1;b1] и [a2; b2]. Напишите программу, которая находит их пересечение.

Пересечением двух отрезков может быть:

отрезок:

точка:

пустое множество:

Формат входных данных
На вход программе подаются четыре целых числа a1,b1,a2,b2 каждое на отдельной строке. Гарантируется, что a1<b1 и a2<b2.

Формат выходных данных
Программа должна вывести на экран границы отрезка, являющегося пересечением, либо общую точку, либо текст «пустое множество».

```python
# Ввод границ первого отрезка
a1 = int(input())
b1 = int(input())

# Ввод границ второго отрезка
a2 = int(input())
b2 = int(input())

# Нахождение пересечения отрезков
left = max(a1, a2)
right = min(b1, b2)

# Проверка наличия пересечения
if left > right:
    print('пустое множество')
elif left == right:
    print(left)
else:
    print(left, right)
```
