```python
import time
start_time = time.perf_counter()

a = [78, -32, 5, 39, 58, -5, -63, 57, 72, 9, 53, -1, 63, -97, -21, -94, -47, 57, -8, 60, -23, -72, -22, -79, 90, 96, -41, -71, -48, 84, 89, -96, 41, -16, 94, -60, -64, -39, 60, -14, -62, -19, -3, 32, 98, 14, 43, 3, -56, 71, -71, -67, 80, 27, 92, 92, -64, 0, -77, 2, -26, 41, 3, -31, 48, 39, 20, -30, 35, 32, -58, 2, 63, 64, 66, 62, 82, -62, 9, -52, 35, -61, 87, 78, 93, -42, 87, -72, -10, -36, 61, -16, 59, 59, 22, -24, -67, 76, -94, 59]

n = len(a)

# реализация алгоритма сортировки выбором

for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if a[j] < a[min_index]:
                min_index = j
        # Меняем местами найденный минимальный элемент с первым элементом этой части
        a[i], a[min_index] = a[min_index], a[i]

print(a)
print(time.perf_counter() - start_time)
```

Сортировка простыми вставками (Insertion sort)

Реализация алгоритма

```python
import time
start_time = time.perf_counter()

a = [1, 7, -3, 9, 0, -67, 34, 12, 45, 1000, 6,  8, -2, 99]
n = len(a)

for i in range(1, n):
    elem = a[i]  # берем первый элемент из неотсортированной части списка
    j = i

    # пока элемент слева существует и больше нашего текущего элемента
    while j >= 1 and a[j - 1] > elem:
        # смещаем j-й элемент отсортированной части вправо
        a[j] = a[j - 1]
        # сами идём влево, дальше ищем место для нашего текущего элемента
        j -= 1

    # нашли место для нащего текущего элемента из неотсортированной части
    # и вставляем его на индекс j в отсортированной части
    a[j] = elem


print('Отсортированный список:', a)
print(time.perf_counter() - start_time)
```

Оптимизация алгоритма

```python
import time

# Функция бинарного поиска для нахождения индекса вставки
def binary_search(arr, val, start, end):
    while start < end:
        mid = (start + end) // 2
        if arr[mid] < val:
            start = mid + 1
        else:
            end = mid
    return start

# Начало замера времени
start_time = time.perf_counter()

a = [1, 7, -3, 9, 0, -67, 34, 12, 45, 1000, 6, 8, -2, 99]
n = len(a)

for i in range(1, n):
    elem = a[i]  # текущий элемент из неотсортированной части списка
    # Находим позицию вставки с использованием бинарного поиска
    pos = binary_search(a, elem, 0, i)
    # Сдвигаем отсортированные элементы вправо, начиная с позиции pos
    a = a[:pos] + [elem] + a[pos:i] + a[i+1:]

# Печать отсортированного списка
print('Отсортированный список:', a)

# Печать времени выполнения
print(time.perf_counter() - start_time)
```

```python
# Минимальный размер подпоследовательности, по умолчанию 32
MIN_MERGE = 32

def calc_min_run(n):
    """Вычисление минимального размера run-length для данного n"""
    r = 0
    while n >= MIN_MERGE:
        r |= n & 1
        n >>= 1
    return n + r
# 1. `n & 1` вычисляет побитовое N для `n` и `1`. Это фактически проверка, является ли `n` нечётным числом. Если `n` нечётное, результат будет `1`. Если чётное — `0`.
#
# 2. `r |= n & 1` присваивает `r` значение побитового ИЛИ текущего `r` и результата выражения `n & 1`. Таким образом, если хотя бы один из `n`(когда оно нечётное) до текущего момента был нечётным, `r` установится в 1.
#
# 3. `n >>= 1` сдвигает `n` вправо на 1 бит (аналогично делению `n` на 2). Этот шаг повторяется до тех пор, пока `n` не станет меньше `MIN_MERGE`.
#
# Когда цикл завершится, значение `r` будет `1`, если хотя бы одно из значений `n` в процессе было нечётным, иначе от останется `0`. В конце функция возвращает сумму последнего `n` и `r`.

def insertion_sort(arr, left, right):
    """Выполняет сортировку вставками на подмассиве arr[left:right+1]"""
    for i in range(left + 1, right + 1):
        key_item = arr[i]
        j = i - 1
        while j >= left and arr[j] > key_item:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item

def merge(arr, l, m, r):
    """Объединяет два отсортированных подмассива arr[l:m+1] и arr[m+1:r+1]"""
    len1, len2 = m - l + 1, r - m
    left, right = [], []
    for i in range(len1):
        left.append(arr[l + i])
    for i in range(len2):
        right.append(arr[m + 1 + i])

    i, j, k = 0, 0, l

    while i < len1 and j < len2:
        if left[i] <= right[j]:
            arr[k] = left[i]
            i += 1
        else:
            arr[k] = right[j]
            j += 1
        k += 1

    while i < len1:
        arr[k] = left[i]
        k += 1
        i += 1

    while j < len2:
        arr[k] = right[j]
        j += 1
        k += 1

def tim_sort(arr):
    n = len(arr)
    min_run = calc_min_run(n)

    for start in range(0, n, min_run):
        end = min(start + min_run - 1, n - 1)
        insertion_sort(arr, start, end)

    size = min_run
    while size < n:
        for left in range(0, n, 2 * size):
            mid = min(n - 1, left + size - 1)
            right = min(n - 1, mid + size)
            if mid < right:
                merge(arr, left, mid, right)
        size *= 2

# Пример использования
arr = [5, 21, 7, 23, 19]
tim_sort(arr)
print(arr)
```
